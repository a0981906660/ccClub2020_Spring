# 讀所有的IP，先讀爬下來的PTT文章

```python
# Save
#np.save("earthquake_dict_np", earthquake_dict)

# Load
earthquake_dict = np.load('earthquake_dict_np.npy',allow_pickle='TRUE').item()
```

```python
#Use fake UserAgent
def set_header_user_agent():
    user_agent = UserAgent()
    return user_agent.random

#open browser #打開瀏覽器，但不要載入圖片
options = webdriver.ChromeOptions()
prefs = {
    'profile.default_content_setting_values': {
        'images': 2,
        #'javascript': 2
    }
}
options.add_experimental_option('prefs', prefs)

#ua = "Mozilla/5.0 (Windows NT 10.0; WOW64; rv:53.0) Gecko/20100101 Firefox/53.0"
#headers = {'User-Agent': 'User-Agent:Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
#user_agent = set_header_user_agent()
#headers={ 'user-agent': user_agent }
#options.add_argument("user-agent={}".format(user_agent)) # 使用偽造的 user-agent
#options.add_argument(headers)

driver = webdriver.Chrome('./chromedriver', chrome_options=options)


for MMMM in earthquake_dict.keys():
    try:
        #print(earthquake_dict[MMMM]["IP"])
        ip_address = earthquake_dict[MMMM]["IP"]
        
        #define urls
        IP_url1 = "https://tw.geoipview.com/"
        IP_url2 = "https://www.ez2o.com/App/Net/IP"
        
        #用第一個反查IP網站
        driver.get(IP_url1)
        driver.implicitly_wait(5)
        
        # 點查詢欄位
        search_field = driver.find_element_by_xpath("""//*[@id="ipinfo_form"]/form/table/tbody/tr/td[2]/table/tbody/tr/td[1]/input""")
        search_field.click()
        search_field.clear()

        # 輸入IP
        search_field.send_keys(ip_address)
        search_field.send_keys(Keys.ENTER)
        driver.implicitly_wait(5)

        #拿到source page
        IP_soup_1 = BeautifulSoup(driver.page_source, 'html.parser')
        driver.implicitly_wait(5)
        
        # 找table
        table_1 = IP_soup_1.find(id = "ipinfo_show")
        df_1 = pd.read_html(str(table_1))[0]
        
        #新增到該筆文章的dictionary下面
        earthquake_dict[MMMM]["Nation"] = df_1.iloc[4][1] #國家
        earthquake_dict[MMMM]["District"] = df_1.iloc[5][1] #地區
        earthquake_dict[MMMM]["City"] = df_1.iloc[6][1] #城市
        earthquake_dict[MMMM]["Latitude"] = df_1.iloc[8][1] #緯度
        earthquake_dict[MMMM]["Longitude"] = df_1.iloc[9][1] #經度
        
        time.sleep(0.5)
    except:
        pass
```
