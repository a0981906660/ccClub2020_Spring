#### 實作時間差

```python
# Set a list that contains 地震文與地震報告的時間差
TimeDifference = []
for j in range(len(df_new)):
    if df_new.iloc[j]["IsOfficialReport"] == 1: #如果是地震報告，則時間差為0
        dstr = df_new.iloc[j]["Date_format"] #找到時間標記
        d1 = datetime.datetime.strptime(dstr, "%Y-%m-%d %H:%M:%S") #Start Time 化為電腦的格式
        TimeDifference.append(0)
    else:
        dstr = df_new.iloc[j]["Date_format"] #找到時間標記
        d2 = datetime.datetime.strptime(dstr, "%Y-%m-%d %H:%M:%S") #地震文的發文時間化為電腦的格式
        d = d2-d1
        TimeDifference.append(d.total_seconds())
```

# 計算距離

台灣地區若以 *北緯24度 東經121度* 即南投埔里附近為原點

則：

* 往東往西差一經度約差101公里

* 往南往北差一經度約差110公里

```python
# Set a list that contains 地震文與地震報告的距離差
Distance = []
for j in range(len(df_new)):
    if df_new.iloc[j]["IsOfficialReport"] == 1: #如果是地震報告，則距離差為0
        X0 = df_new.iloc[j]["Longitude"] #找到經緯度
        Y0 = df_new.iloc[j]["Latitude"]  #當作原點
        Distance.append(0)
    else:
        X = df_new.iloc[j]["Longitude"] #找到經緯度
        Y = df_new.iloc[j]["Latitude"]
        d = np.sqrt( (101*(X-X0))**2 + (110*(Y-Y0))**2 )
        Distance.append(d)
```

```python
df_new["Distance"] = Distance
```

### 尖峰與離峰時段：另外兩個dummies

尖峰時段就設定在peak的前後兩個小時 

* 22:00~2:00

* 13:00~17:00

離峰時間設定在

* 4:00~8:00

* 18:00~20:00

```python
df.iloc[1]["Date_format"]
```

```python
dstr = df.iloc[32]["Date_format"]
d1 = datetime.datetime.strptime(dstr, "%Y-%m-%d %H:%M:%S")
d1.timetuple()
d1.timetuple()[3] #表示23時
```

```python
# Set a list that contains dummies for peak & offpeak
IsPeak = []
OffPeak = []
for j in range(len(df)):
    dstr = df.iloc[j]["Date_format"] #找到時間標記
    d1 = datetime.datetime.strptime(dstr, "%Y-%m-%d %H:%M:%S") #Start Time 化為電腦的格式
    hour = d1.timetuple()[3] #拿到使用者是在幾點發文的
    
    if hour in [22, 23, 0, 1, 13, 14, 15, 16]:
        IsPeak.append(1)
    else:
        IsPeak.append(0)
    
    if hour in [4, 5, 6, 7, 18, 19]:
        OffPeak.append(1)
    else:
        OffPeak.append(0)
print(len(IsPeak), len(OffPeak))
```

```python
df["IsPeak"] = IsPeak
df["OffPeak"] = OffPeak
```
