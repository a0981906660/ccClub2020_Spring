# 合併地震資料及爬文資料

```python
import numpy as np
import pandas as pd
```

```python
df_earthquake = pd.read_csv("earthquake_data_0620.csv")
df_ptt = pd.read_csv("ptt.csv")
```

```python
df_earthquake.columns = ['編號', 'Date_format', '經度', '緯度', '規模', '深度', '位置', '震度']
```

```python
df_earthquake.head()
```

```python
df_ptt.head()
```

## 合併檔案

```python
df = pd.merge(df_earthquake, df_ptt,how='outer', on = "Date_format", sort="Date_format")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.to_csv("combined.csv")
```
