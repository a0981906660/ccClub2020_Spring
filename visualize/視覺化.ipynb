# 標準化製圖流程

```python
names = locals()  # 設定local變數命名每張圖
counter = 0  # 地震計數器
center_pos = [23.583234,120.5825975]  # 為了使地圖置中


for i in range(723): # 等等改回來723
    if df['IsOfficialReport'][i] == 1:  # 公告地震
        counter += 1
        names["fmap_%s" %counter] = folium.Map(location = center_pos, zoom_start = 6.5)
        Popup = 'Seismic intensity: ' + str(df['震度'][i])
        earthquake = folium.Marker(location=[df['Latitude'][i],df['Longitude'][i]], popup=Popup,
                                    icon=folium.Icon(icon='star',color='red'))
        names["fmap_%s" %counter].add_child(child=earthquake)
    else:
        try:
            if df["YouWin"][i] == 1:
                m1 = folium.Marker(location=[df['Latitude'][i],df['Longitude'][i]],
                       popup='<i>Winner</i>',
                       icon=folium.Icon(icon='phone', # Icon類型
                                        color='cadetblue', # Marker顏色
                                        prefix='fa')) # 使用Font Awesome Icons
                names["fmap_%s" %counter].add_child(child=m1)
            else:
                m2 = folium.Marker(location=[df['Latitude'][i],df['Longitude'][i]],
                       icon=folium.Icon(icon='phone', # Icon類型
                                        color='purple', # Marker顏色
                                        prefix='fa')) # 使用Font Awesome Icons
                if m2.location != m1.location:
                    names["fmap_%s" %counter].add_child(child=m2)
        except:
            continue
```

# 觀察台灣地震分佈
* 震度由6到1分別是：紅橙黃綠藍紫

```python
allEarthquake = folium.Map(location = center_pos, zoom_start = 6.5)
for i in range(723):
    if df['IsOfficialReport'][i] == 1:
        if df['震度'][i] == 1:
            earthquake = folium.Marker(location=[df['Latitude'][i],df['Longitude'][i]], popup='<b>center<b>',
                                        icon=folium.Icon(icon='star',color='purple'))
            allEarthquake.add_child(child=earthquake)
        elif df['震度'][i] == 2:
            earthquake = folium.Marker(location=[df['Latitude'][i],df['Longitude'][i]], popup='<b>center<b>',
                                        icon=folium.Icon(icon='star',color='blue'))
            allEarthquake.add_child(child=earthquake)
        elif df['震度'][i] == 3:
            earthquake = folium.Marker(location=[df['Latitude'][i],df['Longitude'][i]], popup='<b>center<b>',
                                    icon=folium.Icon(icon='star',color='green'))
            allEarthquake.add_child(child=earthquake)
        elif df['震度'][i] == 4:
            earthquake = folium.Marker(location=[df['Latitude'][i],df['Longitude'][i]], popup='<b>center<b>',
                                    icon=folium.Icon(icon='star',color='yellow'))
            allEarthquake.add_child(child=earthquake)
        elif df['震度'][i] == 5:
            earthquake = folium.Marker(location=[df['Latitude'][i],df['Longitude'][i]], popup='<b>center<b>',
                                    icon=folium.Icon(icon='star',color='orange'))
            allEarthquake.add_child(child=earthquake)
        elif df['震度'][i] == 6:
            earthquake = folium.Marker(location=[df['Latitude'][i],df['Longitude'][i]], popup='<b>center<b>',
                                    icon=folium.Icon(icon='star',color='red'))
            allEarthquake.add_child(child=earthquake)
```

# 對所有地震建立字典
* {'時間'：地圖}

```python
d = {}
datelist = []
counter = 0
for i in range(723):
    if df['IsOfficialReport'][i] == 1:
        counter += 1
        d[df['Date_format'][i]] = {}
        d[df['Date_format'][i]]['map'] = names["fmap_%s" %counter]
        datelist.append(df['Date_format'][i])
```

```python
def findmap(time):
    return d[time]['map']
```

# 最終版

```python
interact(findmap, time=datelist)
```
