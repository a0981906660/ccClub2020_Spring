```python
#將載好檔案開啟並加入震度、寫入同一個檔案中

with open('C:\\python\\project\\result.csv', 'w', newline='') as outp:
    writer = csv.writer(outp)
    for i in range(1995,2021):
        if i == 2020:
            for j in range(1,6):
                with open(f'C:\\python\\project\\{i}_{j}.csv', newline='') as inp:
                    rows = csv.reader(inp)
                    count = 0
                    for row,index in zip(rows,range(len(result[f'{i}_{j}'])+2)):
                        count += 1
                        if count == 1 or count == 2:
                            continue
                        else:
                            row.append(result[f'{i}_{j}'][index-2])
                            writer.writerow(row)
                inp.close()
        else:
            for j in range(1,13):
                with open(f'C:\\python\\project\\{i}_{j}.csv', newline='') as inp:
                    rows = csv.reader(inp)
                    count = 0
                    for row,index in zip(rows,range(len(result[f'{i}_{j}'])+2)):
                        count += 1
                        if i == 1995 and j == 1:
                            if count == 1:
                                continue
                            elif count == 2:
                                row.append("震度")
                                writer.writerow(row)
                            else:
                                row.append(result[f'{i}_{j}'][index-2])
                                writer.writerow(row)
                        else:
                            if count == 1 or count == 2:
                                continue
                            else:
                                row.append(result[f'{i}_{j}'][index-2])
                                writer.writerow(row)
                    inp.close()
outp.close()
            
```
